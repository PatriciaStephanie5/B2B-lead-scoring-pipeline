```python
import pandas as pd
df = pd.read_csv("01_account_targeting_leads (1).csv")
df.head()
```

```python
df.info()
df.describe()
df['industry'].value_counts()
```

```python
df = df.drop_duplicates()
df = df.dropna()

def score_lead(row):
    score = 0

    if row['industry'] == 'SaaS':
        score += 3
    if row['employee_count'] > 200:
        score += 2
    if row['engagement_score'] > 70:
        score += 5
    return score
df['lead_score'] = df.apply(score_lead, axis=1)
```

```python
df = df.sort_values(by='lead_score', ascending=False)
df.head(10)
```

```python
df.to_csv("scored_leads.csv", index=False)
```
